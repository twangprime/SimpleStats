```python
import pandas as pd
import numpy as np
import json

#Input the data 
input_file = './inflationdata.txt'
df = pd.read_csv(
    input_file,
    sep='\t'
    )

#Strip out spaces in column names 
df.columns = df.columns.str.replace(' ','')

df=df[['year','value']]

df=df.head(10000)
df['value']=df['value'].astype(float)

df_year = df.groupby(by=["year"]).mean()

df_year['value']=df_year['value'].round(2)

df_year=df_year.reset_index()


```

```python
df_year
```

```python
#Convert to JSON: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_json.html
result = df_year.to_json(orient="records") #[1:-1].replace('},{', '} {')

#Parse the JSON so that it doesn't contain unnecessary backslashes
parsed = json.loads(result)
json.dumps(parsed, indent=4)

```

```python
import json
with open('inflationdata.json', 'w') as f:
    json.dump(parsed, f)
```
